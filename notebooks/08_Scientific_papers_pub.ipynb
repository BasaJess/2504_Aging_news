```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data with the correct delimiter
df = pd.read_csv("../data/Scientific_pub_per_million.csv", delimiter=';')

# Rename columns for convenience
df.columns = ['year', 'publications_per_million']

# Convert 'year' to numeric, if it's not already
df['year'] = pd.to_numeric(df['year'], errors='coerce')

# Convert 'publications_per_million' to numeric, handling commas as decimal separators
# Convert 'publications_per_million' to string before replacing commas
df['publications_per_million'] = df['publications_per_million'].astype(str).str.replace(',', '.', regex=False)

df['publications_per_million'] = pd.to_numeric(df['publications_per_million'], errors='coerce')



# Drop any rows with missing values
df.dropna(inplace=True)

# Ensure data is sorted by year
df = df.sort_values('year')

# Extract x and y values
x = df['year']
y = df['publications_per_million']

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data Points')

# Compute linear regression (best-fit line)
coefficients = np.polyfit(x, y, 1)
slope, intercept = coefficients
best_fit_line = slope * x + intercept

# Plot best-fit line
plt.plot(x, best_fit_line, color='red', label=f'Best Fit Line: y = {slope:.2f}x + {intercept:.2f}')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Publications per Million People')
plt.title('Scientific Publications per Year')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data with the correct delimiter and handle potential header issues
df = pd.read_csv("../data/Scientific_pub_per_million.csv", delimiter=';', header=0)

# Check if the DataFrame has only one column due to improper parsing
if df.shape[1] == 1:
    # Split the single column into two using the semicolon delimiter
    df = df.iloc[:, 0].str.split(';', expand=True)
    df.columns = ['year', 'publications_per_million']
else:
    # Rename columns for convenience
    df.columns = ['year', 'publications_per_million']

# Convert 'year' to numeric, if it's not already
df['year'] = pd.to_numeric(df['year'], errors='coerce')

# Convert 'publications_per_million' to numeric, handling commas as decimal separators
df['publications_per_million'] = df['publications_per_million'].astype(str).str.replace(',', '.', regex=False)
df['publications_per_million'] = pd.to_numeric(df['publications_per_million'], errors='coerce')

# Drop any rows with missing values
df.dropna(inplace=True)

# Ensure data is sorted by year
df = df.sort_values('year')

# Extract x and y values
x = df['year']
y = df['publications_per_million']

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data Points')

# Fit a polynomial of degree 3 (cubic)
degree = 3
coefficients = np.polyfit(x, y, degree)
polynomial = np.poly1d(coefficients)

# Generate x values for plotting the polynomial curve
x_curve = np.linspace(x.min(), x.max(), 500)
y_curve = polynomial(x_curve)

# Plot the polynomial curve
plt.plot(x_curve, y_curve, color='red', label=f'Best Fit Curve (Degree {degree})')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Publications per Million People')
plt.title('Scientific Publications per Year')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
